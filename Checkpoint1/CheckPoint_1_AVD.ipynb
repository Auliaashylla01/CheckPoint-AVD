##Business Objective

Dataset Loan Approval

##Assess Situation

Dataset ini berisi informasi finansial dan demografis dari 1.000 pemohon pinjaman, termasuk pendapatan tahunan, skor kredit, jumlah pinjaman, jumlah tanggungan, status pekerjaan, dan status persetujuan.
